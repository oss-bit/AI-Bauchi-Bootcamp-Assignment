### Exploring canning edges for edge detection
 - Using various images and threshold values to explore the effects of canning edges on edge detection

```python
#Reading the images 

import cv2

first_image = cv2.imread("Images/first.jpg")
second_image = cv2.imread("Images/second.jpg")
third_image = cv2.imread("Images/third.jpeg")

resized_first_image = cv2.resize(first_image, (200,200))
resized_second_image = cv2.resize(second_image, (200,200))
resized_third_image = cv2.resize(third_image, (200,200))
```

```python
#Declaring threshold values and exprimeting with 250 as values for thresholding
threshold_1 = 250
threshold_2 = 250

#getting the edges for the various images
first_image_edges = cv2.Canny(resized_first_image, threshold_1, threshold_2)
second_image_edges = cv2.Canny(resized_second_image, threshold_1, threshold_2)
third_image_edges = cv2.Canny(resized_third_image, threshold_1, threshold_2)
```

```python
#displaying the resulting edges

import matplotlib.pyplot as plt

#configuring the figure size
plt.figure(figsize=(200,10))

#displaying the Original images
plt.subplot(2,3,1)
plt.imshow(resized_first_image)
plt.title("Original First Image")
plt.axis("off")
plt.show()

plt.subplot(2,3,2)
plt.imshow(resized_second_image)
plt.title("Original Second Image")
plt.axis("off")
plt.show()

plt.subplot(2,3,3)
plt.imshow(resized_third_image)
plt.title("Original Third Image")
plt.axis("off")
plt.show()


#displaying the edges of the images

plt.subplot(2,3,4)
plt.imshow(first_image_edges)
plt.title("First Image Edges")
plt.axis("off")
plt.show()

plt.subplot(2,3,5)
plt.imshow(second_image_edges)
plt.title(" Second Image Edges")
plt.axis("off")
plt.show()

plt.subplot(2,3,6)
plt.imshow(third_image_edges)
plt.title("Thrid Image Edges")
plt.axis("off")
plt.show()
```

### Trying the 200 and 250 as threshold values

```python
#trying the threshold of 200 and 250

def try_threshold(threshold_1, threshold_2):

    first_image_edges = cv2.Canny(resized_first_image, threshold_1, threshold_2)
    second_image_edges = cv2.Canny(resized_second_image, threshold_1, threshold_2)
    third_image_edges = cv2.Canny(resized_third_image, threshold_1, threshold_2)

    plt.figure(figsize=(200,10))

    #displaying the Original images
    plt.subplot(2,3,1)
    plt.imshow(resized_first_image)
    plt.title("Original First Image")
    plt.axis("off")
    plt.show()

    plt.subplot(2,3,2)
    plt.imshow(resized_second_image)
    plt.title("Original Second Image")
    plt.axis("off")
    plt.show()

    plt.subplot(2,3,3)
    plt.imshow(resized_third_image)
    plt.title("Original Third Image")
    plt.axis("off")
    plt.show()


    #displaying the edges of the images

    plt.subplot(2,3,4)
    plt.imshow(first_image_edges)
    plt.title("First Image Edges")
    plt.axis("off")
    plt.show()

    plt.subplot(2,3,5)
    plt.imshow(second_image_edges)
    plt.title(" Second Image Edges")
    plt.axis("off")
    plt.show()

    plt.subplot(2,3,6)
    plt.imshow(third_image_edges)
    plt.title("Thrid Image Edges")
    plt.axis("off")
    plt.show()

try_threshold(200, 250)
```

### Trying 100 and 150 as threshold values

```python
try_threshold(100, 150)
```

### Trying for the threshold values of 50 100

```python
try_threshold(50, 100)
```

### Observation
-  As the values for both threshold get smaller the canny Edge algorithm detects more edges
