### Feature detection with SIFT

```python
#import the needed libararies
import cv2
import matplotlib.pyplot as plt

#Loading the images
First_image =  cv2.imread("Images/atsushi1.jpg")
Second_image = cv2.imread("Images/atsushi2.jpg")

First_gray_image = cv2.cvtColor(First_image, cv2.COLOR_BGR2GRAY)
Second_gray_image = cv2.cvtColor(Second_image, cv2.COLOR_BGR2GRAY)

sift = cv2.SIFT_create()
first_keypoints, _ = sift.detectAndCompute(First_gray_image, None)
second_keypoints, _ = sift.detectAndCompute(Second_gray_image, None)

first_sift_image = cv2.drawKeypoints(First_image, first_keypoints, None)
second_sift_image = cv2.drawKeypoints(Second_image, second_keypoints, None)


plt.figure(figsize=(10, 10))
plt.subplot(2,1,1)
plt.imshow(cv2.cvtColor(first_sift_image, cv2.COLOR_BGR2RGB))
plt.title("first image with sift keypoints")
plt.axis("off")
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(2,1,2)
plt.imshow(cv2.cvtColor(second_sift_image, cv2.COLOR_BGR2RGB))
plt.title("second image with sift keypoints")
plt.axis("off")
plt.show()

```

```python
print(len(first_keypoints))
print(len(second_keypoints))
```

### Feature detection with SURF

```python
sift = cv2.SIFT_create()

# Detect keypoints and descriptors
first_keypoints = sift.detect(First_gray_image,None)
second_keypoints = sift.detect(Second_gray_image,None)

# Draw keypoints on the image
first_surf_image = cv2.drawKeypoints(First_gray_image,first_keypoints,First_image)
second_surf_image = cv2.drawKeypoints(Second_gray_image,second_keypoints,Second_image)

# Display the result using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(first_surf_image, cv2.COLOR_BGR2RGB))
plt.title('SURF Keypoints')
plt.axis("off")
plt.show()

plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(second_surf_image, cv2.COLOR_BGR2RGB))
plt.title('SURF Keypoints')
plt.axis("off")
plt.show()
```

```python
print(len(first_keypoints))
print(len(second_keypoints))
```

### Feature detection with ORB

```python
# Creating the ORB detector object
orb = cv2.ORB_create()

# detect keypoints and descriptors
first_keypoints, _ = orb.detectAndCompute(First_gray_image, None)
second_keypoints, _ = orb.detectAndCompute(Second_gray_image, None)

# drawing keypoints on the images
first_orb_image = cv2.drawKeypoints(First_image, first_keypoints, None)
second_orb_image = cv2.drawKeypoints(Second_image, first_keypoints, None)

# Display the result using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(first_orb_image, cv2.COLOR_BGR2RGB))
plt.title('First Image with ORB Keypoints')
plt.axis('off')
plt.show()

plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(second_orb_image, cv2.COLOR_BGR2RGB))
plt.title('Second Image with ORB Keypoints')
plt.axis('off')
plt.show()
```

```python
print(len(first_keypoints))
print(len(second_keypoints))
```
